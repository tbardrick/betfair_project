## Expected value

It is useful to know the expected value of a hedge for similating/backtesting strategies.

Uncertainty will always remain in the final payoff of a hedge bet, because:
1. It is possible that the second position of the hedge strategy isn't matched
2. If the position isnt matched, the payoff is subject to whether the horse wins
3. The price of the second positon is unknown (although we know that we impose an upper/lower boundary)

### New concepts: p(win), p(match)

We need the probabiity of a match, and the probability of a win. The probability of a win implied by the price is the best proxy available for the probability of a win (given by $\frac{1}{p}$).

### Expected Value formula for back-to-lay

The expected value formula differs based on hedge type. For back-to-lay:

$E[\pi] = p(match) \cdot \pi_H + [1 - p(match)][p(win)\cdot \pi_{back|win} + (1 - p(win))\cdot \pi_{back | loss}]$

where 
* $p(win) = \frac{1}{p}$
* the payoff from a succesful hedge $\pi_H = (bp - 1) \cdot bs \cdot (1 - c) - (lp - 1) \cdot ls = - bs + ls \cdot (1 - c)$
* the payoff from winning back bet $\pi_{back|win} = (bp - 1) \cdot bs \cdot (1 - c) $
* the payoff (loss) from losing back bet $\pi_{back|loss} = - bs$
