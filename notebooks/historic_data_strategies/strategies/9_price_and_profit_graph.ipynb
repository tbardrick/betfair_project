### Section 0 : Setup

```python
# importing packages
from pathlib import Path, PurePath 

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import utils
```

```python
# reading in data
project_dir = Path.cwd().parents[2]
data_dir = project_dir / 'data' / 'processed' / 'api' / 'advanced' / 'adv_data.csv'
df = pd.read_csv(data_dir, index_col = 0)
print(df.shape)
```

```python
# defining variables
back_prices = [col for col in df.columns if 'BP' in col]
back_sizes = [col for col in df.columns if 'BS' in col]
lay_prices = [col for col in df.columns if 'LP' in col]
lay_sizes = [col for col in df.columns if 'LS' in col]

runner_info = ['SelectionId', 'MarketId', 'Venue', 'Distance', 'RaceType', 'BSP', 'NoRunners']
```

```python
def plot_df():
    plot_df = df[df['BSP'].notna()].sample(1).copy() # 60 NA BSP's

    bp_df = plot_df[['SelectionId'] + back_prices].copy()
    bp_df.columns = bp_df.columns.str.replace("[BP:T]", "")
    bp_df.columns = bp_df.columns.str.replace("[+]", "")
    bp_df = pd.melt(bp_df, id_vars = ['SelectionId'])
    bp_df['variable'] = bp_df['variable'].astype(int)

    lp_df = plot_df[['SelectionId'] + lay_prices].copy()
    lp_df.columns = lp_df.columns.str.replace("[LP:T]", "")
    lp_df.columns = lp_df.columns.str.replace("[+]", "")
    lp_df = pd.melt(lp_df, id_vars = ['SelectionId'])
    lp_df['variable'] = lp_df['variable'].astype(int)

    bsp_df = plot_df[['BSP']].copy()
    
    prof_bp_df_gt0 = bp_df.copy()
    prof_bp_df_gt0['profit'] = prof_bp_df_gt0['value'].apply(lambda x: utils.payout_lay_to_back(x, bsp_df['BSP'].item(), 10, 0.05))
    prof_bp_df_gt0 = prof_bp_df_gt0.loc[prof_bp_df_gt0['variable'] > 0].copy()
    
    prof_lp_df_gt0 = lp_df.copy()
    prof_lp_df_gt0['profit'] = prof_lp_df_gt0['value'].apply(lambda x: utils.payout_back_to_lay(bsp_df['BSP'].item(), 10, x, 0.05))
    prof_lp_df_gt0 = prof_lp_df_gt0.loc[prof_lp_df_gt0['variable'] > 0].copy()
    
    prof_bp_df_lt0 = bp_df.copy()
    prof_bp_df_lt0['profit'] = prof_bp_df_lt0['value'].apply(lambda x: utils.payout_back_to_lay(x, 10, bsp_df['BSP'].item(), 0.05))
    prof_bp_df_lt0 = prof_bp_df_lt0.loc[prof_bp_df_lt0['variable'] < 0].copy()
    
    prof_lp_df_lt0 = lp_df.copy()
    prof_lp_df_lt0['profit'] = prof_lp_df_lt0['value'].apply(lambda x: utils.payout_lay_to_back(bsp_df['BSP'].item(), x, 10, 0.05))   
    prof_lp_df_lt0 = prof_lp_df_lt0.loc[prof_lp_df_lt0['variable'] < 0].copy()
    
    return bp_df, lp_df, bsp_df, prof_bp_df_gt0, prof_lp_df_gt0, prof_bp_df_lt0 , prof_lp_df_lt0
```

However, what we are most interest in is how to extract most value from this insight by quanitfying how often this occurs - we need some metrics...

__2.2 Metrics__

To determine the potential success of this strategy we need to consider three things:
    - (1) The probability of the inplay price reaching a profitable level
    - (2) The probability of your inplay bet being matched
    - (3) The profit that can be made if prices reaches/exceeds this level
    - (4) The loss incurred if the price doesn't reach this level and the trade isn't match
    
The strategy would look something like this:
    - We back and lay at BSP 
    - We place seperate lay and back orders into the market as the race starts
    - The orders are placed at the minimum price differencee from the BSP needed to make a profitable hedge
    
(We are effectively testing out two strategies at once, Back to then Lay and Lay to then Back)

To get an initial idea of the potential success of this strategy we'll initially just investigate the following:
    - How often does an inplay back price exceed the BSP?
    - How often does an inplay lay price go below the BSP?
    
It could also be intersting to know the inplay price variance for each runner (could this occur multiple times per runner in a race, does it happen a lot to some but not to other). To get an idea of this we'll also calculate the variance...

- variance of inplay BP/LP per runner
    
    
Now this is a good start, but we're really interest in is the amount by which these prices increase/decrease by and if these differences are profitable enough...
    
- if inplay BP went above bsp? does it go higher than min hedge price?
- if inplay LP went below bsp? does it go higher than min hedge price?
